## Visualización de datos

*Para identificar relaciones en la información (información, patrones, resumenes, lagunas, ruidos y atípicos)*

**Scatter plot**

Ver como dependen las relaciones entre ellos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Generar datos simulados
np.random.seed(42)
num_pozos = 50

profundidad = np.random.randint(800, 4000, num_pozos)  # Profundidad del pozo (m)
presion = profundidad * np.random.uniform(1.5, 2.0, num_pozos)  # Presión del yacimiento (psi)

# Crear un DataFrame
data = pd.DataFrame({'Profundidad (m)': profundidad, 'Presión (psi)': presion})

# Crear scatter plot con Seaborn
plt.figure(figsize=(8,6))
sns.regplot(
    x=data['Profundidad (m)'], 
    y=data['Presión (psi)'], 
    scatter_kws={'color': 'blue', 'alpha': 0.6},  # Puntos azules con transparencia
    line_kws={'color': 'red'},  # Línea de regresión roja
    ci=None  # Sin intervalo de confianza
)

# Etiquetas y título
plt.xlabel("Profundidad del Pozo (m)")
plt.ylabel("Presión del Yacimiento (psi)")
plt.title("Dependencia entre Profundidad y Presión en un Pozo Petrolero")

# Mostrar el gráfico
plt.show()

```

*A mayor profundidad, mayor presión.  
La línea de regresión resalta esta tendencia lineal.  
La pendiente de la línea nos da una idea del gradiente de presión, es decir, cuánto aumenta la presión por cada metro de profundidad.  
Este tipo de análisis es crucial para la planificación y evaluación de pozos, ya que conocer cómo se comportan la profundidad y la presión ayuda a diseñar equipos adecuados y gestionar los riesgos asociados al desarrollo de los pozos.*

**Gráfico de barras**

El objetivo es ver como están organizados con respecto a algo, cuántos objetos caen en ciertas áreas o categorías.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Datos simulados: Producción de petróleo (barriles/día) por pozo
pozos = ['Pozo A', 'Pozo B', 'Pozo C', 'Pozo D', 'Pozo E']
produccion = [1200, 1500, 1000, 1300, 1100]  # Barriles por día

# Crear un gráfico de barras simple
plt.figure(figsize=(8,6))
sns.barplot(x=pozos, y=produccion, palette='Blues')

# Etiquetas y título
plt.xlabel("Pozo Petrolero")
plt.ylabel("Producción de Petróleo (barriles/día)")
plt.title("Producción de Petróleo por Pozo")

# Mostrar el gráfico
plt.show()
```

 Podemos comparar visualmente la producción de cada pozo, mostrando claramente cuáles son los pozos más productivos y cuáles menos.

Este tipo de gráfico es útil para visualizar rendimientos de producción, comparaciones entre pozos, o para hacer informes de rendimiento en operaciones petroleras.

**Diagrama de barras usando una variable adicional**

Sirve para comparar ciertos indicadores, como se relacionan 2 variables entre ellos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Datos simulados: Producción de petróleo (barriles/día) por pozo en dos formaciones productoras
pozos = ['Pozo A', 'Pozo B', 'Pozo C', 'Pozo D', 'Pozo E']
formaciones = ['Formación 1', 'Formación 2']
# Generamos datos aleatorios de producción para cada pozo y formación
produccion_data = {
    'Pozo': ['Pozo A', 'Pozo A', 'Pozo B', 'Pozo B', 
             'Pozo C', 'Pozo C', 'Pozo D', 'Pozo D', 
             'Pozo E', 'Pozo E'],
    'Formación': ['Formación 1', 'Formación 2', 'Formación 1', 'Formación 2',
                  'Formación 1', 'Formación 2', 'Formación 1', 'Formación 2',
                  'Formación 1', 'Formación 2'],
    'Producción (barriles/día)': [1200, 1500, 1300, 1600, 
                                  1100, 1400, 1350, 1450,
                                  1250, 1550]
}

# Crear un DataFrame
data = pd.DataFrame(produccion_data)

# Crear gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(x='Pozo', y='Producción (barriles/día)', hue='Formación', data=data)

# Etiquetas y título
plt.xlabel("Pozo Petrolero")
plt.ylabel("Producción de Petróleo (barriles/día)")
plt.title("Comparación de Producción de Petróleo entre Formaciones por Pozo")

# Mostrar el gráfico
plt.show()



```

Este gráfico de barras comparativas te permite ver claramente cómo la producción de petróleo de cada pozo varía entre las dos formaciones productoras. Al comparar las alturas de las barras de cada pozo, puedes identificar rápidamente cuál formación produce más petróleo en cada uno de los pozos, lo que puede ser útil para optimizar la extracción en función de la productividad de las formaciones.

**Grafico de Pastel**

Apreciar un panorama general de los datos en porcentaje.

```python
import matplotlib.pyplot as plt

# Datos simulados: Distribución de formaciones geológicas en un conjunto de pozos
formaciones = ['Arenisca', 'Caliza', 'Dolomita', 'Esquisto']
# Número de pozos perforados en cada tipo de formación
cantidad_formaciones = [6, 3, 5, 2]  # Número de pozos por tipo de formación

# Crear gráfico de pastel
plt.figure(figsize=(8,8))
plt.pie(cantidad_formaciones, labels=formaciones, autopct='%1.1f%%', startangle=90, colors=['#ffb3e6', '#ff6666', '#99ccff', '#66b3ff'])

# Título del gráfico
plt.title("Distribución de Formaciones Geológicas en Pozos Petroleros")

# Mostrar gráfico
plt.show()

```

Muestra la distribución de formaciones geológicas en un conjunto de pozos petroleros. Cada porción del pastel indica el porcentaje de pozos perforados en cada tipo de formación geológica. Este tipo de gráfico es útil para comprender cómo se distribuyen los esfuerzos de perforación y si alguna formación geológica es predominante en la exploración petrolera.
